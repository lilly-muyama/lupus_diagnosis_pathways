```python
import pandas as pd
import numpy as np
import random
```

```python
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
```

```python
# adding missingness function
def insert_nans(df, col, frac):
    non_null_indices = df[df[col].notna()].index.tolist() #getting not null indices
    to_change_indices = random.sample(non_null_indices, int(frac*len(non_null_indices)))
    df.loc[to_change_indices, col] = np.nan
    return df    
```

```python
# the train set
train_df = pd.read_csv('../../data/train_set_noisiness_0.2.csv')
train_df.head()
```

```python
train_df.isna().sum()
```

```python
# example missingness fraction
frac = 0.1
```

```python
for column in train_df.columns:
    if column not in ['ana', 'label']:
        train_df = insert_nans(train_df, column, frac)
train_df.isna().sum()
```
