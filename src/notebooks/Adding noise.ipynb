```python
import pandas as pd
import numpy as np
import random
```

```python
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
```

```python
# the training set
train_df = pd.read_csv('../../data/train_set_basic.csv')
train_df.head()
```

```python
train_df.label.value_counts()
```

```python
def insert_noise(train_df, frac):
    '''
    Adds noise to the dataset
    '''
    noisy_train_df = train_df.copy(deep=True)
    idx_list = noisy_train_df.sample(frac=frac, random_state=SEED).index.tolist()
    for idx in idx_list:
        if noisy_train_df.loc[idx]['label'] == 0:
            noisy_train_df.at[idx, 'label'] = 1
        elif noisy_train_df.loc[idx]['label'] == 1:
            noisy_train_df.at[idx, 'label'] = 0
    return noisy_train_df
```

```python
# example noise level fraction
frac = 0.1
```

```python
noisy_train_df = insert_noise(train_df, frac)
noisy_train_df.label.value_counts()
```
