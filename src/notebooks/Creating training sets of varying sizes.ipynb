```python
import pandas as pd
import numpy as np
import random
```

```python
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
```

```python
train_df = pd.read_csv('../../data/train_set_basic.csv')
train_df.head()
```

```python
fracs = [0.01, 0.05, 0.1, 0.5]
```

```python
def sample_train_set(df, frac):
    sample_num = int(frac*len(df))
    idx_list = random.sample(list(df.index), sample_num)
    sampled_df = df.loc[idx_list]
    sampled_df = sampled_df.reset_index(drop=True)
#     sampled_df.to_csv(f'../../data/train_set_basic_{frac}_seed_{SEED}.csv', index=False)
```

```python
for frac in fracs:
    sample_train_set(train_df, frac)
```
